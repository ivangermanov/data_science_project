```python
df_orders = pd.read_csv(orders_path,
                        #parse_dates=['DoB', 'Order date'],
                        dtype=
                        {'Customer number': np.int64,
                         'Gender': object,
                         'PoR': object,
                         'Order number': np.int64,
                         'Product number': object,
                         'Sub category': object,
                         'Category': object,
                         'Count': np.int16,
                         'Price': np.float64,
                         'EDT': np.int16,
                         'ADT': np.int16,
                         'RoR': object,
                         'Rating': np.int8})
```

```python
df_ordered = df_orders.groupby(['Category', 'Sub category'])
print(df_ordered[['Category', 'Sub category']].nunique().to_string())
```

```python
df_categories = df_orders.groupby(['Category'])['Price'].sum()
df_categories
```

```python
df_orders['Order date'] = pd.to_datetime(df_orders['Order date'])
df_orders.tail(5)
```

```python
df_orders['day_of_week'] = df_orders['my_dates'].dt.day_name()
df_orders.tail(5)
```

```python
df_orders.groupby(["day_of_week"])["day_of_week"].agg("count").sort_values(ascending=False)
```
