## Libraries

```python
import numpy as np
import pandas as pd

import os
import matplotlib.pyplot as plt

%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

## Dataset

```python
orders_path = '../orders_after_cleaning.csv'
df_orders = pd.read_csv(orders_path)
```

```python
df_orders = df_orders.drop(df_orders.columns[0], axis=1)
```

## Count the orders per every customer

```python
df_orders_per_customer = df_orders.groupby('Customer number')['Customer number', 'Order number']
df_orders_per_customer.head(10)
```

## Plot distribution of orders

```python
sns.set_style('darkgrid')
x = pd.Series(df_orders_per_customer['Order number'], name="Order count")
plt.subplots(figsize=(15,10))
sns.distplot(x)
plt.title("Distribution of orders")
print()
```
